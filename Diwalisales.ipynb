```python
                                              #---------------Exploratory data analysis--------------#
```

```python
ax=srn.countplot(x='Gender',data=df) 
#for labeling above the container
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
#Barplot through seaborn
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
srn.barplot(x='Gender',y='Amount',data= sales_gen)
```

```python
ax=srn.countplot(data=df,x='Age Group',hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
srn.barplot(x='Age Group',y='Amount',data= sales_age,hue='Gender')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
srn.set(rc={"figure.figsize":(15,5)})
srn.barplot(x='State',y='Orders',data= sales_state)
```

```python
ax=srn.countplot(data=df,x= 'Marital_Status')
```

```python
sales_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
srn.set(rc={"figure.figsize":(6,5)})
srn.barplot(x='Marital_Status',y='Amount',data= sales_status,hue='Gender')
```

```python
srn.set(rc={"figure.figsize":(20,5)})
ax=srn.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu=df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
srn.set(rc={"figure.figsize":(20,5)})
srn.barplot(x='Occupation',y='Amount',data= sales_occu)
```

```python
srn.set(rc={"figure.figsize":(20,5)})
ax=srn.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu=df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
srn.set(rc={"figure.figsize":(20,5)})
srn.barplot(x='Product_Category',y='Amount',data= sales_occu)
```
